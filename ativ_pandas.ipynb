# LER DATASET

```python
url_path = "https://storage.googleapis.com/dados_brutos_soulcode/original/Estacoes_Geral.csv"
```

```python
col_dates = ['Data Inclusão','Data Primeiro Licenciamento','Data Último Licenciamento', 'Data Validade Licença']
```

```python
pd.set_option('max_columns', None)
```

```python
df = pd.read_csv(url_path, sep=';', parse_dates=col_dates, dayfirst=True)
```

```python
# criação do Schema de dados

schema = pa.DataFrameSchema(
    columns = {
        "identifição":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "cidade":pa.Column(pa.String),
        "uf":pa.Column(pa.String,pa.Check.str_length(2,2),nullable=True),
        "aerodromo":pa.Column(pa.String,pa.Check.str_length(4,4),nullable=True),
        "data":pa.Column(pa.DateTime),
        "horario":pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9]):([0-5][0-9])?$'),nullable=True),
        "liberado":pa.Column(pa.String,pa.Check.str_length(3,3),nullable=True),
        "status":pa.Column(pa.String,nullable=True),
        "total_recomendacoes":pa.Column(pa.Int),
        "total_aeronaves":pa.Column(pa.Int),
        "saida_pista":pa.Column(pa.String,pa.Check.str_length(3,3))

    }
)
```
